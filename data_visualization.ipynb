```python
import pandas as pd

forest_fires = pd.read_csv("forestfires.csv")
forest_fires.head(5)
```

** Making a scatter plot with the wind column on the x-axis and the area column on the y-axis: **

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(forest_fires['wind'], forest_fires['area'])
plt.title("Wind speed vs. fire area")
plt.ylabel("Area consumed by fire")
plt.xlabel("Wind speed when fire started")
plt.show()
```

** Making a scatter plot with the temp column on the x-axis and the area column on the y-axis: **

```python
plt.scatter(forest_fires['temp'], forest_fires['area'])
plt.title("Temperature in Celsius vs. Fire area")
plt.ylabel("Area consumed by fire")
plt.xlabel("Temperature in Celsius when fire started")
plt.show()
```

** We've used the pivot_table() method to create the area_by_y and area_by_x dataframes. Each dataframe contains the average area of the fires started at each X or Y position in forest_fires. By plotting this out, we'll be able to see which areas in the park are the most flammable. **

** Because the x-axis has to contain numeric values, we'll use the index property of area_by_y and area_by_x. This is because the X and Y columns are numeric, so the index values will also be numeric. **

```python
import numpy
area_by_y = forest_fires.pivot_table(index="Y", values="area", aggfunc=numpy.mean)
area_by_x = forest_fires.pivot_table(index="X", values="area", aggfunc=numpy.mean)

plt.bar(area_by_y.index, area_by_y)
plt.show()
plt.bar(area_by_x.index, area_by_x)
plt.show()
```

** We've used the pivot_table() method to create the area_by_month and area_by_day dataframes. Each dataframe contains the average area of the fires started for different month and day values in forest_fires. By plotting this out, we'll be able to see how large the fires get at different times of the year.**

** We'll need to make an extra step here before we plot. Because the index properties of area_by_month and area_by_day contain strings (month and day are strings), we'll need to use range(len(area_by_month)) and range(len(area_by_day)) on the y-axis instead of area_by_month.index and area_by_day.index. The range() function will generate a sequence from 1 to the integer that is passed into it. **

```python
area_by_month = forest_fires.pivot_table(index="month", values="area", aggfunc=numpy.mean)
area_by_day = forest_fires.pivot_table(index="day", values="area", aggfunc=numpy.mean)

plt.barh(range(len(area_by_month)), area_by_month)
plt.show()

plt.barh(range(len(area_by_day)), area_by_day)
plt.show()
```

** Now plot a line chart: **

```python
#Grouping temperatures when fire started by month and calculating the average:
temp_by_month = forest_fires.pivot_table(index="month", values="temp", aggfunc=numpy.mean)

plt.plot(range(len(temp_by_month)), temp_by_month)
plt.title("Average montlhy temperature variation")
plt.xlabel("Month")
plt.ylabel("Temperature in Celsius")
plt.show()
```

** Playing around with styles: **

```python
## The FiveThirtyEight style:
plt.style.use("fivethirtyeight")
plt.plot(range(len(temp_by_month)), temp_by_month)
plt.title("Average montlhy temperature variation")
plt.xlabel("Month")
plt.ylabel("Temperature in Celsius")
plt.show()

## The ggplot style:
plt.style.use("ggplot")
plt.plot(range(len(temp_by_month)), temp_by_month)
plt.title("Average montlhy temperature variation")
plt.xlabel("Month")
plt.ylabel("Temperature in Celsius")
plt.show()

## The dark_background style:
plt.style.use("dark_background")
plt.plot(range(len(temp_by_month)), temp_by_month)
plt.title("Average montlhy temperature variation")
plt.xlabel("Month")
plt.ylabel("Temperature in Celsius")
plt.show()

## The bmh style:
plt.style.use("bmh")
plt.plot(range(len(temp_by_month)), temp_by_month)
plt.title("Average montlhy temperature variation")
plt.xlabel("Month")
plt.ylabel("Temperature in Celsius")
plt.show()

```

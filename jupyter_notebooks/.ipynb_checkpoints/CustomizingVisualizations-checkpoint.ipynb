```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")

recent_grads.head()
```

```python
recent_grads_droppedNA = recent_grads.dropna()
recent_grads_droppedNA.shape[0]
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.hist(recent_grads_droppedNA['ShareWomen'])
ax2.scatter(recent_grads_droppedNA['Unemployment_rate'], recent_grads_droppedNA['ShareWomen'])
ax3.scatter(recent_grads_droppedNA['ShareWomen'], recent_grads_droppedNA['Unemployment_rate'])
ax4.hist(recent_grads_droppedNA['Unemployment_rate'])

ax1.xaxis.set_visible(False)
ax2.xaxis.set_visible(False)

ax1.set_ylabel("ShareWomen")
ax3.set_xlabel("ShareWomen")
ax4.set_xlabel("Unemployment_rate")

ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)
ax4.get_yaxis().set_visible(False)

plt.subplots_adjust(wspace=False, hspace=False)

ax1.set_ylim(0,30)
ax2.set_ylim(0.0,1.0)
ax3.set_xlim(0.0,1.0)
ax3.set_ylim(0.0,0.20)
ax4.set_xlim(0.0,0.20)

ax1.set_yticklabels([0, 5, 10, 15, 20, 25, 30])
ax3.set_yticklabels([0.00, 0.05, 0.10, 0.15])
ax3.set_xticklabels([0.0, 0.2, 0.4, 0.6, 0.8], rotation=90)
ax4.set_xticklabels([0.00, 0.05, 0.10, 0.15, 0.20], rotation=90)

plt.show()
```

```python
recent_grads_droppedNA['ShareMen'] = recent_grads_droppedNA['Men'] / recent_grads_droppedNA['Total']
```

```python
arts = recent_grads_droppedNA[recent_grads_droppedNA["Major_category"] == "Arts"]
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

import numpy as np
n = len(arts["Major"].unique())
locs = np.arange(0,n)
bar_width = 0.35
offset_locs = locs + bar_width

bar_1 = ax.bar(left=locs, height=arts['ShareMen'].tolist(), width=bar_width)
bar_2 = ax.bar(left=offset_locs, height=arts['ShareWomen'].tolist(), width=bar_width, color="green")

ax.set_xticklabels(arts["Major"].tolist(), rotation=90)

ax.set_xticks(offset_locs)

plt.legend((bar_1, bar_2), ("ShareMen", "ShareWomen"),loc="upper left")
plt.grid()
plt.show()
```

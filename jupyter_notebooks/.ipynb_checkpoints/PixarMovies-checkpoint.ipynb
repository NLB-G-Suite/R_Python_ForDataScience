** Pixar Movies Data Analysis and Visualization Project ** 

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

%matplotlib inline

pixar_movies = pd.read_csv("PixarMovies.csv")

print("Loading Pixar movies dataset:")
print("# Rows:",pixar_movies.shape[0])
print("# Columns:",pixar_movies.shape[1])
```

Just 15 rows (15 movies)... So let's see them all:

```python
pixar_movies.head(15)
```

```python
print("Columns datatypes:")
pixar_movies.dtypes
```

```python
print("Basic descriptive statistics:")
pixar_movies.describe()
```

** Two columns (Domestic % and International %) have their numerical values with the '%' char at the end. This makes the columns become as object type instead of float type. **
So... Let's remove the % char from the end of each value in those columns.

```python
pixar_movies['Domestic %'] = pixar_movies['Domestic %'].str.rstrip('%').astype("float")
```

Let's check the result:

```python
pixar_movies.dtypes
```

Ok. Now let's do the same with the "International %" column

```python
pixar_movies['International %'] = pixar_movies['International %'].str.rstrip('%').astype("float")
```

** Now making the scores columns in the same scale: **

The IMDB Score column is on a 10 point scale while the RT Score and Metacritic Score columns are on a 100 point scale. We need to convert the IMDB scores to a 100 point scale so we can compare how the movies fared across these different review sites.

```python
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score'] * 10
```

Checking this last change:

```python
pixar_movies.head()
```

** Removing the Inside Out movie: ** 

The last row, describing the movie Inside Out, is missing values for Oscars Nominated and Oscars Won. This is because the dataset is only upto date as of November 2015 and doesn't contain the Oscar nomations or results. Let's avoid using data on Inside Out for any plots involving revenue figures or the Oscars since those values aren't complete yet.

```python
filtered_pixar = pixar_movies.loc[0:13]
```

Let's see the filtered dataset:

```python
filtered_pixar.head(14)
```

** Movie titles as index: ** 

To finish out our data preparation, let's set the movie title as a dataset index, so we can access the rows by referring to the movie titles.

```python
filtered_pixar.set_index("Movie", drop=False, inplace=True)
print("My favorite Pixar movie is:\n",filtered_pixar.loc["WALL-E"])
```

** Let's answer questions with just visualizing the data **

First: "How do the Pixar films fare across each of the major review sites?"

Let's generate the following 3 line plots:
   * Rotten Tomatoes score on the y-axis, movie name on the x-axis.
   * IMDB score on the y-axis, movie name on the x-axis.
   * Metacritic score on the y-axis, movie name on the x-axis.

```python
#Subseting to extract only the scores columns
critics_reviews = filtered_pixar[['RT Score','IMDB Score','Metacritic Score']]
```

```python
#Let's see how it goes with just a simple line plotting without adjustments
critics_reviews.plot()
#plt.show()
```

The resulting chart is a little cramped, so let's tweak the figsize parameter and increase the size.

```python
critics_reviews.plot(figsize=(15,6))
```

** Well... According to the plot the review site Rotten Tomatoes gives Pixar consistently higher ratings. Let's generate a box plot to explore the question: **

How are the average ratings from each review site across all the movies distributed?

```python
critics_reviews.plot(kind="box", figsize=(9,5))
```

Fine... These boxplots confirm what we've noticed: Rotten Tomatoes scores are higher in average. Also, we can se better using the boxplots that the Metacritic Scores vary more, which means that their values are more spread out (ranging from under 60 to under 100).

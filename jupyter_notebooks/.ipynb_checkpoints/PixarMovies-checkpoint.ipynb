** Pixar Movies Data Analysis and Visualization Project ** 

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

%matplotlib inline

pixar_movies = pd.read_csv("PixarMovies.csv")

print("Loading Pixar movies dataset:")
print("# Rows:",pixar_movies.shape[0])
print("# Columns:",pixar_movies.shape[1])
```

Just 15 rows (15 movies)... So let's see them all:

```python
pixar_movies.head(15)
```

```python
print("Columns datatypes:")
pixar_movies.dtypes
```

```python
print("Basic descriptive statistics:")
pixar_movies.describe()
```

** Two columns (Domestic % and International %) have their numerical values with the '%' char at the end. This makes the columns become as object type instead of float type. **
So... Let's remove the % char from the end of each value in those columns.

```python
pixar_movies['Domestic %'] = pixar_movies['Domestic %'].str.rstrip('%').astype("float")
```

Let's check the result:

```python
pixar_movies.dtypes
```

Ok. Now let's do the same with the "International %" column

```python
pixar_movies['International %'] = pixar_movies['International %'].str.rstrip('%').astype("float")
```

** Now making the scores columns in the same scale: **

The IMDB Score column is on a 10 point scale while the RT Score and Metacritic Score columns are on a 100 point scale. We need to convert the IMDB scores to a 100 point scale so we can compare how the movies fared across these different review sites.

```python
pixar_movies['IMDB Score'] = pixar_movies['IMDB Score'] * 10
```

Checking this last change:

```python
pixar_movies.head()
```

** Removing the Inside Out movie: ** 

The last row, describing the movie Inside Out, is missing values for Oscars Nominated and Oscars Won. This is because the dataset is only upto date as of November 2015 and doesn't contain the Oscar nomations or results. Let's avoid using data on Inside Out for any plots involving revenue figures or the Oscars since those values aren't complete yet.

```python
filtered_pixar = pixar_movies.loc[0:13]
```

Let's see the filtered dataset:

```python
filtered_pixar.head(14)
```

** Movie titles as index: ** 

To finish out our data preparation, let's set the movie title as a dataset index, so we can access the rows by referring to the movie titles.

```python
pixar_movies.set_index("Movie", drop=False, inplace=True)
filtered_pixar.set_index("Movie", drop=False, inplace=True)
print("My favorite Pixar movie is:\n",filtered_pixar.loc["WALL-E"])
```

** Let's answer questions with just visualizing the data **

First: "How do the Pixar films fare across each of the major review sites?"

Let's generate the following 3 line plots:
   * Rotten Tomatoes score on the y-axis, movie name on the x-axis.
   * IMDB score on the y-axis, movie name on the x-axis.
   * Metacritic score on the y-axis, movie name on the x-axis.

```python
#Subseting to extract only the scores columns
critics_reviews = pixar_movies[['RT Score','IMDB Score','Metacritic Score']]
```

```python
#Let's see how it goes with just a simple line plotting without adjustments
critics_reviews.plot()
#plt.show()
```

The resulting chart is a little cramped, so let's tweak the figsize parameter and increase the size.

```python
critics_reviews.plot(figsize=(15,6))
```

** Well... According to the plot the review site Rotten Tomatoes gives Pixar consistently higher ratings. Let's generate a box plot to explore the question: **

How are the average ratings from each review site across all the movies distributed?

```python
critics_reviews.plot(kind="box", figsize=(9,5))
```

Fine... These boxplots confirm what we've noticed: Rotten Tomatoes scores are higher in average. Also, we can se better using the boxplots that the Metacritic Scores vary more, which means that their values are more spread out (ranging from under 60 to under 100).

** How has the ratio of where the revenue comes from changed since the first movie? Now that Pixar is more well known internationally, is more revenue being made internationally for newer movies? **

Answering using a ** Stacked bar plot: **

```python
revenues_filtered_pixar = filtered_pixar[['Domestic %', 'International %']]
revenues_filtered_pixar.plot(kind='bar', stacked=True)
```

The proportion of domestic revenue has slightly decreased while the proportion of international revenue has slightly increased.

** Now let's just visualy analyze whether there is a correlation or not between scores and oscars won **

```python
critics_and_oscars = critics_reviews.copy(deep=True)
critics_and_oscars['Oscars Won'] = filtered_pixar['Oscars Won']
critics_and_oscars.plot(figsize=(15,6))
```

Ooops... As you can see, the scores vary much more and in a different scale. There isn't many oscars for Pixar movies.
So one possible action is to change the scores scale to a 0 to 10 scale.

```python
critics_and_oscars['RT Score'] = critics_and_oscars['RT Score'] / 10
critics_and_oscars['IMDB Score'] = critics_and_oscars['IMDB Score'] / 10
critics_and_oscars['Metacritic Score'] = critics_and_oscars['Metacritic Score'] / 10
critics_and_oscars.plot(figsize=(15,6))
```

** Some conclusions: **
   * Overall, critics review and oscars do not seem to correlate.
   * The first thre movies had great scores but did not win oscars. 
   * After Toy Story 2, Pixar won oscars with high score movies until Cars.
   * It seems that the movive Cars has disappointed the fans with an explicit drop in the scores and also no oscars won.
   * There are two scores drop breaking a sequence of oscars won. So this is an important observation. 

** Although there is not a strong and apparent correlation, according to this data there is a likelihood to keep winning oscars if there is not a drop on critics review. So great reviews do not necessarily implies on Oscars won, however if Pixar is in a sequence of winning Oscars, a drop on critics review may somehow imply in breaking the sequence of winning Oscars. **

Of course this is a shallow analysis. For example, we could add more features and apply machine learning to build a predictive model and test the accuracy of the model with and without critics scores in predicting Oscars won.

** Now turning to domestic gross. Let's see what correlates with "Adjusted Domestic Gross", which describes the total domestic revenue adjusted for economic and ticket price inflation **

```python
adg_opening_budget = filtered_pixar[['Adjusted Domestic Gross', 'Opening Weekend', 'Production Budget']]
adg_opening_budget.plot(kind='line', figsize=(15,6))
```

Opening weekend revenue seems to be correlated with production budget. However, overall they are not correlated with adjusted domestic gross. We can only see for certain movies a slight correlation between adjusted domestic gross and production budget.

Let's check with the critics scores.

```python
adg_critics = filtered_pixar[['Adjusted Domestic Gross', 'RT Score', 'IMDB Score', 'Metacritic Score']]
adg_critics.plot(kind='line', figsize=(15,6))
```

There seems to be a correlation. However, it is hard to see due to differences on scale. Let's normalizing them applyin min-max scaling.

```python
from sklearn import preprocessing
minmax_scale = preprocessing.MinMaxScaler().fit(adg_critics)
df_minmax = minmax_scale.transform(adg_critics)
adg_critics.loc[:,'Adjusted Domestic Gross'] = df_minmax[:,0]
adg_critics.loc[:,'RT Score'] = df_minmax[:,1]
adg_critics.loc[:,'IMDB Score'] = df_minmax[:,2]
adg_critics.loc[:,'Metacritic Score'] = df_minmax[:,3]
adg_critics.plot(kind='line', figsize=(15,6))
```

** Now we can better see that there's a correlation between critics scores and Adjusted Domestic Gross (ADG). There are only two exceptions: (i) After Cars, the critics score increased while the ADG decreased; and, (ii) After WALL-E the critics score decreased whereas the ADG increased. **
